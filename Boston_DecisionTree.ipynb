```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
boston = pd.read_csv(r'C:\Users\Thota Divya\Desktop\INtellipaat---SQL\Python datasets\python Datasets\Boston.csv')
```

```python
boston.head()
```

<h2>Dropping Unnamed: 0	column as it is of no use in model building or analysis

```python
boston.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
boston.head()
```

<h2> Visualising the plot 'rm' i.e average rooms per dwelling vs 'medv' i.e median value of price in $1000s

```python
plt.scatter(data=boston,x = 'rm',y='medv',color = 'green')
plt.xlabel('Average rooms per dwelling')
plt.ylabel('Median value of price in $1000s')
plt.title('Average rooms per dwelling vs Median value of price in $1000s')
plt.show()
```

<h2>Getting the feature and target from the dataset

```python
x = pd.DataFrame(boston['rm'])
y = pd.DataFrame(boston['medv'])
```

<h2>Splitting the data into training and test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

<h2>Building the model

```python
regressor = DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

<h2>Predicting the values

```python
y_pred = regressor.predict(x_test)
```

<h2>Finding the rmse value

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred,y_test)
rmse = np.sqrt(mean_squared_error(y_pred,y_test))
rmse
```

<h1>Till now we were working on 1 features/independent variable lets work on multiple features now

```python
x = pd.DataFrame(boston[['rm','lstat']])# features/independent variables
y = pd.DataFrame(boston['medv'])# target                 
```

<h2>Splitting the data into training and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20)
```

<h2>Building the model

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

<h2>Predicting the values

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred,y_test)
rmse = np.sqrt(mse)
rmse
```

<h1>Insights:

```python
Observed that the model error is less when 'rm','lstat' are used as features for predicting the 'medv' to that of using 'rm' alone as feature
```
