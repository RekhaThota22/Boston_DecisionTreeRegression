```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
# lstat = population below poverty line
# medv = median price
# First lets do a simple linear regression using 1 x and 1 y
data = df.loc[:,['lstat','medv']]
data.head()
```

```python
# Visualizing the dta
data.plot(x='lstat',y = 'medv',style = 'o')
data.xlabel('lstat')
data.ylabel('medv')
```

# Preparing the model:

```python
X = pd.DataFrame(data['lstat'])
Y = pd.DataFrame(data['medv'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)

y_pred = lr.predict(X_test)
y_pred = pd.DataFrame(y_pred,columns = ['predicted'])

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print(f"Mean Squared Error : {mean_squared_error(Y_test,y_pred)}")
print(f"Mean Absolute Error : {mean_absolute_error(Y_test,y_pred)}")
print(f"R2 Score : {r2_score(Y_test,y_pred)}")
print(f'Root Mean sqyared Error: {np.sqrt(mean_squared_error(Y_test,y_pred))}')
```

# Multiple Linear Regression

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
X = pd.DataFrame(df.loc[:,df.columns != 'medv'])
Y = pd.DataFrame(df.loc[:,df.columns == 'medv'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)

y_pred = lr.predict(X_test)
y_pred = pd.DataFrame(y_pred,columns = ['predicted'])

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print(f"Mean Squared Error : {mean_squared_error(Y_test,y_pred)}")
print(f"Mean Absolute Error : {mean_absolute_error(Y_test,y_pred)}")
print(f"R2 Score : {r2_score(Y_test,y_pred)}")
print(f'Root Mean sqyared Error: {np.sqrt(mean_squared_error(Y_test,y_pred))}')
```
